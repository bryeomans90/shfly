```python
def explore_customer_date(ex_custno):
    
    one_cust_online = online[online['custno']==ex_custno].copy()
    one_cust_order = order[order['custno']==ex_custno].copy()

    # Clean up date time
    one_cust_order['dt'] = one_cust_order['orderdate']
    one_cust_full = pd.concat([one_cust_online,one_cust_order]).sort_values('dt')
    display(one_cust_full)
```

```python
for id in test_customer_ids:
    print(f'Customer ID:{id}')
    explore_customer_date(id)
```
